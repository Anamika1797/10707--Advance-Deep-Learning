```python
import numpy as np
loss1=[]
with open('loss1.txt') as f_out:
    lines = f_out.readlines()
    for line in lines:
        key=line.replace("\n", "")
        print(key)
        key=key.split(",")
        loss1.append(float(key[1]))
        #print(line[-2])
```

```python
loss2=[]
with open('loss2.txt') as f_out:
    lines = f_out.readlines()
    for line in lines:
        key=line.replace("\n", "")
        print(key)
        key=key.split(",")
        loss2.append(float(key[1]))
        #print(line[-2])
```

```python
import matplotlib.pyplot as plt
plt.plot(range(50),loss1,label='With RBM')
plt.plot(range(50),loss2,label='Without RBM')
plt.title('Loss Vs Epoch for hidden unit=100 and k=5')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
